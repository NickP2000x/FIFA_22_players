```python
import pandas as pd
import numpy as np
```

### clean up on fifa players dataset (going to do regression on rating and value of player)

```python
dat = pd.read_csv("Fifa 22 Fut Players.csv")
dat.head()
```

```python
dat.info()
```

### data cleaning 

```python
#check na and duplicates, make new features from body column (height, weight, body type) and value 
```

```python
dat.isnull().sum()
```

```python
dat[dat["Body"].isnull()].head()
```

```python
dat.dropna(inplace=True)
```

```python
dat.drop(dat[dat["PS"] == '0'].index, inplace=True)
```

```python
dat.sample(10)
```

```python
dat["Height(cm)"] = dat.Body.str.extract('(\d+)cm')
dat["Weight(kg)"] = dat.Body.str.extract('(\d+)kg').astype(np.int32)
dat["Body Type"] = dat.Body.str.extract('" (.+) \(')
```

```python
dat[["Height(cm)", "Body Type"]].isnull().sum()
```

```python
dat = dat.dropna()
```

```python
dat[["Height(cm)", "Body Type"]].isnull().sum()
```

```python
#prevents warning 
pd.options.mode.chained_assignment = None  # default='warn'
dat["Height(cm)"] = dat["Height(cm)"].astype(np.int32)
```

```python
dat["Height(cm)"].dtypes
```

```python
dat.isnull().sum()
```

```python
dat.drop(columns="Body", inplace=True)
```

```python
ps_ext = dat.PS.str.extract('(\d+[\.]?\d*)([MK]?)').dropna()
ps_ext[0] = ps_ext[0].astype(np.float32)
ps_ext
```

```python
new_nums = []
for x, y, z in ps_ext.itertuples():
  if z == 'M':
    new_nums.append(round(y, ndigits=2) * 1000000)
  elif z == 'K':
    new_nums.append(round(y, ndigits=2) * 1000)
  else:
    new_nums.append(y)
```

```python
new_ps = pd.Series(new_nums)
```

```python
dat["PS"] = new_ps
```

```python
dat['Body Type'] = dat['Body Type'].astype('string')
dat["Name"] = dat["Name"].astype('string')
```

```python
dat.dropna(inplace=True)
```

```python
dat.to_csv("fifa_22_fut_players_cleaned.csv")
```
